```python
from FastExplain import *
import random
import pandas as pd
```

```python
feature_1 = [i  for i in range(10)]*100
feature_2 = [i/2 + 2*random.random() for i in feature_1]
feature_3 = [i/2 + 2*random.random() for i in feature_2]
uncorrelated_feature = [random.random() for i in feature_1]
y = [0.1*i + j  for i,j in zip(feature_1, uncorrelated_feature)]
df = pd.DataFrame({'feature_2': feature_2, 'feature_1': feature_1, 'feature_3': feature_3, 'uncorrelated_feature': uncorrelated_feature, 'y': y})
```

```python
df.head()
```

# XGBoost

## Assigns correlated attribution to a single feature

```python
xgb = model_data(df, 'y', model='xgb')
```

```python
xgb.plot_ale('feature_1')
```

```python
xgb.plot_ale('feature_2')
```

```python
xgb.plot_ale('feature_3')
```

```python
xgb.plot_ale('uncorrelated_feature')
```
